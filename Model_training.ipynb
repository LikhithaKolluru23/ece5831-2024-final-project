```python
import json
import os
from glob import glob

# Paths
json_dir = "/content/drive/MyDrive/COCO_dataset/labels/train"  # Directory containing your JSON files
output_dir = "/content/drive/MyDrive/COCO_dataset/labels/train1"  # Directory to save YOLO .txt files (same structure as json_dir)

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define category mapping: original category_id -> YOLO class_id
category_mapping = {
    18: 0,  # Dog
    42: 1,  # Surfboard
    38: 2,  # Kite
    11: 3,  # Fire Hydrant
    9: 4,   # Boat
    51: 5,  # Bowl
    46: 6,  # Wine Glass
    39: 7,  # Baseball Bat
    25: 8,  # Giraffe
    17: 9,  # Cat
}

def convert_bbox_to_yolo(size, bbox):
    """Convert COCO bbox format to YOLO format."""
    img_width, img_height = size
    x_min, y_min, width, height = bbox
    x_center = x_min + width / 2.0
    y_center = y_min + height / 2.0
    return [
        x_center / img_width,
        y_center / img_height,
        width / img_width,
        height / img_height,
    ]

# Iterate through all JSON files
for json_file in glob(f"{json_dir}/*.json"):
    with open(json_file, "r") as f:
        annotations = json.load(f)

    # Output file name (replace .json with .txt)
    output_file = os.path.join(output_dir, os.path.basename(json_file).replace(".json", ".txt"))

    with open(output_file, "w") as out_f:
        # Process each annotation
        for annotation in annotations:
            category_id = annotation["category_id"]
            bbox = annotation["bbox"]
            img_width, img_height = (640, 640)  # Replace with actual image dimensions

            # Map category_id to YOLO class_id
            if category_id not in category_mapping:
                print(f"Skipping unrecognized category_id: {category_id}")
                continue  # Skip if category_id is not in the mapping

            yolo_class_id = category_mapping[category_id]

            # Convert bbox to YOLO format
            yolo_bbox = convert_bbox_to_yolo((img_width, img_height), bbox)

            # Create YOLO annotation line
            yolo_annotation = f"{yolo_class_id} " + " ".join(map(str, yolo_bbox)) + "\n"
            out_f.write(yolo_annotation)

    print(f"Converted {json_file} to {output_file}")

print("All files converted to YOLO format!")
```

```python
from ultralytics import YOLO

# Load YOLO model
model = YOLO("yolov8n.pt")  # Load a pretrained YOLO model

# Train YOLO model
model.train(
    data="data.yaml",  # Path to your YAML file
    epochs=25,         # Number of epochs
    imgsz=640          # Image size
)
```

```python
img = "/content/drive/MyDrive/COCO_dataset/train/images/000000000072.jpg"
results = model.predict(img)

unique_labels = set()
for result in results:
    unique_labels.update([model.names[int(cls)] for cls in result.boxes.cls])

print("Unique Labels:", unique_labels)
```

```python
import os

def count_files(directory, extension):
    return len([f for f in os.listdir(directory) if f.endswith(extension)])

train_images = "/content/drive/MyDrive/COCO_dataset/train/images"
train_labels = "/content/drive/MyDrive/COCO_dataset/train/labels"

val_images = "/content/drive/MyDrive/COCO_dataset/val/images"
val_labels = "/content/drive/MyDrive/COCO_dataset/val/labels"

print("Train images:", count_files(train_images, ".jpg"))
print("Train labels:", count_files(train_labels, ".txt"))

print("Val images:", count_files(val_images, ".jpg"))
print("Val labels:", count_files(val_labels, ".txt"))
```
